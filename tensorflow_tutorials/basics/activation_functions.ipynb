# Activaction functions
- Activation function is a function that decide the output of a neuron based on the inputs and weighted values from prior layer's neurons. 

### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Step function
- Step function is a very beginning and basic activation function where the total summed input value (include multiplying weighted value) exceeds a certain threshold, it returns an output value of 1, or 0 otherwise.

```python
def step(x):
    return np.array(x > 0, dtype=np.int32)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)

plt.plot(x, y, color='b')
plt.title('Step function')
plt.show()
```

### Sigmoid function

- Sigmoid function is used frequently in a binary classification model in order to decide whether the output is true or false, 0 or 1 and spam or not spam.

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 1.0], linestyle=':', color='r')
plt.title('Sigmoid function')
plt.show()
```

### Hyperbolic tangent function

- Hyperbolic tangent function returns the output value in between -1 to 1. It is also often used for a binary classification.

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y, color='b')
plt.axhline(color='g', linestyle='--')
plt.plot([0, 0], [-1.0, 1.0], linestyle=':', color='r')
plt.title('Hyperbolic tangent function')
plt.show()
```

### Rectified linear unit (ReLU)

- The most popular activation function is rectified linear unit, also called the ReLU. ReLU function has a unique characteristic. It returns the output of 0 whenever the input is negative (include 0), and when the input is positive, the ReLU function returns the output same as input value.

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 5.0], linestyle=':', color='r')
plt.title('Rectified linear unit (ReLU)')
plt.show()
```

### Softmax function

- Softmax function, unlike sigmoid and hyperbolic tangent function that are used in a binary classification, is implemented to classify a certain value from three or more categories. It is called as the multiclass classification.
- The input values can be positive, negative, zero, or greater than one, but the softmax activation function transforms them into values between 0 and 1, so that they can be interpreted as probabilities. (Reference: https://deepai.org/machine-learning-glossary-and-terms/softmax-layer)

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y, color='b')
plt.title('Softmax function')
plt.show()
```
