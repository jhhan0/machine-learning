## Different activaction functions in neutral networks

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Step function

```python
def step(x):
  return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)

plt.plot(x, y, 'r')
plt.title('Step function')
plt.show()
```

### Sigmoid function

```python
def sigmoid(x):
  return (1 / (1 + np.exp(-x)))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 1.0], linestyle=':', color='r')
plt.title('Sigmoid function')
plt.show()
```

### Hyperbolic tangent function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y, color='b')
plt.axhline(color='g', linestyle='--')
plt.plot([0, 0], [-1.0, 1.0], linestyle=':', color='r')
plt.title('Hyperbolic tangent function')
plt.show()
```

### Rectified linear unit (ReLU)

```python
def relu(x):
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 5.0], linestyle=':', color='r')
plt.title('Rectified linear unit (ReLU)')
plt.show()
```
