## Different activaction functions in neutral networks

**Activation function is a function that decide the output of a neuron based on the inputs and weighted values from prior layer's neurons.** 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Step function

**Step function is a very beginning and basic activation function where the total summed input value (include multiplying weighted value) exceeds a certain threshold, it returns an output value of 1, or 0 otherwise.**

```python
def step(x):
  return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)

plt.plot(x, y, color='b')
plt.title('Step function')
plt.show()
```

### Sigmoid function

**Sigmoid function is used frequently in a binary classification model in order to decide whether the output is true or false, 0 or 1, spam or not spam.**

```python
def sigmoid(x):
  return (1 / (1 + np.exp(-x)))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 1.0], linestyle=':', color='r')
plt.title('Sigmoid function')
plt.show()
```

### Hyperbolic tangent function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y, color='b')
plt.axhline(color='g', linestyle='--')
plt.plot([0, 0], [-1.0, 1.0], linestyle=':', color='r')
plt.title('Hyperbolic tangent function')
plt.show()
```

### Rectified linear unit (ReLU)

```python
def relu(x):
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y, color='b')
plt.plot([0, 0], [0, 5.0], linestyle=':', color='r')
plt.title('Rectified linear unit (ReLU)')
plt.show()
```

### Softmax function

**Softmax function, unlike sigmoid and hyperbolic tangent function used in a binary classification, is implemented to classify a certain value from three or more categories. It is called the multiclass classification.**

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y, color='b')
plt.title('Softmax function')
plt.show()
```
